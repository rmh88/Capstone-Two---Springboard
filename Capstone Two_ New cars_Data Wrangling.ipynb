```python
# Look at nan values and further explore data and features to decide how to address them

df_cars.isna().sum()
```

```python
df_cars[df_cars['MSRP'].isna()]
```

```python
df_cars['MSRP'] = df_cars['MSRP'].fillna('Missing')
df_cars.shape
```

```python
df_cars['EPA Fuel Economy Est - City (MPG)'].unique()
```

```python
df_cars['EPA Fuel Economy Est - City (MPG)'] = df_cars['EPA Fuel Economy Est - City (MPG)'].str.split().str[0]
df_cars['EPA Fuel Economy Est - City (MPG)'] = df_cars['EPA Fuel Economy Est - City (MPG)'].apply(pd.to_numeric, errors = 'coerce')
df_cars['EPA Fuel Economy Est - City (MPG)'].unique()
```

```python
df_cars.isna().sum()
```

```python
print(np.nanmedian(df_cars['EPA Fuel Economy Est - City (MPG)']))
print(np.mean(df_cars['EPA Fuel Economy Est - City (MPG)']))
```

```python
df_cars['EPA Fuel Economy Est - City (MPG)'].fillna(np.nanmedian(df_cars['EPA Fuel Economy Est - City (MPG)']), inplace = True)
```

```python
col_list = ['Engine', 'EPA Classification', 'Drivetrain', 'Body Style', 'Transmission', 'Fuel System', 'Trans Description Cont.']
df_cars[col_list] = df_cars[col_list].fillna('Missing')
```

```python
col_list2 = ['Brake Type', 'Steering Type', 'Suspension Type - Front', 'Suspension Type - Rear', 'Front Wheel Material']
df_cars[col_list2] = df_cars[col_list2].fillna('Missing')
df_cars.isna().sum()
```

```python
df_cars['Drivetrain Years'].value_counts()
```

```python
df_cars['Turning Diameter - Curb to Curb (ft)'].unique()
```

```python
df_cars['Roadside Assistance Years'].unique()
```

```python
df_cars['Roadside Assistance Miles/km'].unique()
```

```python
col_list = ['Base Curb Weight (lbs)', 'Height, Overall (in)', 'Fuel Tank Capacity, Approx (gal)', 'Trans Type', 'Turning Diameter - Curb to Curb (ft)']

df_cars[col_list] = df_cars[col_list].apply(pd.to_numeric, errors = 'coerce')
df_cars.dtypes
```

```python
print(np.nanmedian(df_cars['Passenger Volume (ftÂ³)']))
print(np.nanmedian(df_cars[['Passenger Volume (ftÂ³)', 'Wheelbase (in)', 'Track Width, Front (in)']]))
      
print(df_cars['Passenger Volume (ftÂ³)'].median())
print(df_cars[['Passenger Volume (ftÂ³)', 'Wheelbase (in)', 'Track Width, Front (in)']].median())
```

```python
col_list3 = ['Passenger Volume (ftÂ³)', 'Wheelbase (in)', 'Track Width, Front (in)', 'Basic Years', 'Corrosion Years'] + col_list

df_cars[col_list3] = df_cars[col_list3].fillna(value = df_cars[col_list3].median())

print(df_cars[['Passenger Volume (ftÂ³)', 'Wheelbase (in)', 'Track Width, Front (in)']].median())
print(df_cars[col_list3].isna().sum())
```

```python
df_cars.isna().sum()
```

```python
col_list4 = ['SAE Net Torque @ RPM', 'SAE Net Horsepower @ RPM', 'Displacement', 'Basic Miles/km', 'Corrosion Miles/km', 'Drivetrain Miles/km', 'Drivetrain Years', 'Front Tire Size', 'Roadside Assistance Years', 'Roadside Assistance Miles/km']

df_cars[col_list4] = df_cars[col_list4].fillna('Missing')
df_cars.isna().sum()
```

```python
# Decided to drop Stabilizer bar Diameter column since 2/3rds of all values are NaN

df_cars.drop(df_cars.columns[[-3]], axis = 1, inplace = True)
df_cars.head(20)
```
